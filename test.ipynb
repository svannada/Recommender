```python
import pandas as pd
import imdb
import time

# Initialize IMDb access
access = imdb.IMDb()

# Function to get cover URL for a movie ID
def get_cover_url(imdb_id):
    try:
        # Convert the IMDb ID to an integer
        imdb_id = int(imdb_id)
        movie = access.get_movie(imdb_id)
        return movie.get('cover url')
    except Exception as e:
        print(f"Error fetching cover URL for IMDb ID {imdb_id}: {e}")
        return None

def add_cover_urls_to_csv(csv_path, output_csv_path):
    # Load movie links CSV
    movie_links = pd.read_csv(csv_path)

    # Add cover URL column
    movie_links['cover_url'] = None
    
    # Track number of requests made
    request_count = 0

    # Iterate over rows and fetch cover URLs
    for idx, row in movie_links.iterrows():
        imdb_id = row['imdbId']
        cover_url = get_cover_url(imdb_id)
        movie_links.at[idx, 'cover_url'] = cover_url
        request_count += 1

        # Rate limit to avoid hitting IMDb server too frequently
        if request_count % 10 == 0:
            time.sleep(1)  # Sleep for 1 second every 10 requests

    # Save to output CSV
    movie_links.to_csv(output_csv_path, index=False)

```

```python
# Example usage
add_cover_urls_to_csv('ml-latest-small/links.csv', 'ml-latest-small/links_with_urls.csv')
```
